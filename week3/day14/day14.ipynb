```python
import numpy as np
import pandas as pd

np.random.seed(10)

sales_data = pd.DataFrame({
    "Day": np.arange(1,31),
    "Sales": np.random.normal(50000, 8000, 30)
})

sales_data
```

```python
mean = sales_data["Sales"].mean()
median = sales_data["Sales"].median()
mode = sales_data["Sales"].mode()[0]

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
np.random.seed(20)

salary_data = pd.DataFrame({
    "Salary": np.random.normal(60000,15000,200)
})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(y=salary_data["Salary"])
plt.title("Salary Distribution")
plt.show()
```

```python
np.random.seed(30)

waiting_time = pd.DataFrame({
    "Waiting_Time": np.random.exponential(20,150)
})
```

```python
variance = waiting_time["Waiting_Time"].var()
range_val = waiting_time["Waiting_Time"].max() - waiting_time["Waiting_Time"].min()

print("Variance:", variance)
print("Range:", range_val)
```

```python
Q1 = waiting_time["Waiting_Time"].quantile(0.25)
Q3 = waiting_time["Waiting_Time"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = waiting_time[
    (waiting_time["Waiting_Time"] < lower) |
    (waiting_time["Waiting_Time"] > upper)
]

print("Outliers:\n", outliers)
```
