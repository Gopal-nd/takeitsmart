```python
import numpy as np
import matplotlib.pyplot as plt

days = np.arange(1,31)
stock_price = np.random.normal(500,20,30)

plt.plot(days,stock_price,marker='o',color='b')
plt.title("Stock Price Movement")
plt.xlabel("Days")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
traffic = np.random.randint(1000,5000,30)

plt.plot(days,traffic,color='g')
plt.title("Website Traffic Trend")
plt.show()
```

```python
salary = np.random.normal(60000,15000,200)

plt.hist(salary,bins=20,color='c')
plt.title("Salary Distribution")
plt.show()
```

```python
defaults = [300,700]
labels = ["Default","Non-Default"]

plt.pie(defaults,labels=labels,autopct='%1.1f%%')
plt.title("Loan Default Rate")
plt.show()
```
