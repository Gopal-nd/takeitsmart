```python
import numpy as np
import pandas as pd

np.random.seed(10)

sales_30 = pd.DataFrame({
    "Day": np.arange(1,31),
    "Sales": np.random.normal(50000, 8000, 30)
})

sales_30
```

```python
mean = sales_30["Sales"].mean()
median = sales_30["Sales"].median()
mode = sales_30["Sales"].mode()[0]

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
std = sales_30["Sales"].std()
var = sales_30["Sales"].var()

print("Standard Deviation:", std)
print("Variance:", var)
```

```python
sales_30["Ad_Spend"] = np.random.normal(20000, 3000, 30)

correlation = sales_30[["Sales","Ad_Spend"]].corr()
print(correlation)
```
