```python
import pandas as pd

# Create DataFrame
data = {
    "name": ["Ajay", "Kavya", "Rahul", "Sneha", "Vikram"],
    "dept": ["IT", "HR", "IT", "Finance", "HR"],
    "salary": [60000, 45000, 75000, 80000, 50000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Filter by department (IT)
it_dept = df[df["dept"] == "IT"]
print("\nIT Department:")
print(it_dept)

# Group by department
grouped = df.groupby("dept")["salary"].mean()
print("\nAverage Salary by Department:")
print(grouped)
```
