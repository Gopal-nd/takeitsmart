```python
File_name='grades.txt'
def calculate_grade(avg):
    if avg>=90:
        return 'A'
    elif avg>=75:
        return 'B'
    elif avg>=60:
        return 'C'
    elif avg>=40:
        return 'D'
    else:
        return 'F'

def add_student():
    name=input("Enter Student name:").strip()
    if not name:
        print("Name cannot be empty")
        return

    try:
        marks=[]
        for i in range(1,4):
            mark=int(input(f"Enter mark{i}:"))
            if mark<0 or mark>100:
                raise ValueError
            marks.append(mark)

        total=sum(marks)
        average=round(total/len(marks),2)
        grade=calculate_grade(average)

        with open(File_name, "a") as file:
          file.write(
             f"{name},{marks[0]},{marks[1]},{marks[2]},{total},{average},{grade}\n"
    )
        print("student records added successfully")


    except ValueError:
        print("please enter valid marks between 0 and 100")

def view_students():
    try:
        with open(File_name,"r") as file:
            print("\n Student grade book")
            print("-")

            for line in file:
                name,m1,m2,m3,total,avg,grade=line.strip().split(",")
                print(f"Name:{name}")
                print(f"Marks:[{m1},{m2},{m3}]")
                print(f"Total:{total}")
                print(f"Average:{avg}")
                print(f"Grade:{grade}")
                print("-")

    except FileNotFoundError:
        print("no records found")

def main():
    while True:
        print("\n Student Grade Book Manager")
        print("1. Add student")
        print("2. view students")
        print("3. Exit")

        choice=input("Enter choice:")

        if choice=="1":
            add_student()
        elif choice=="2":
            view_students()
        elif choice=="3":
           print("Exit the program")
        else:
            print("Invalid choice")
```
