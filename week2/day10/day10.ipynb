```python
import numpy as np

days = 30

# 1️⃣ Daily Customers (Poisson)
daily_customers = np.random.poisson(lam=200, size=days)

# 2️⃣ Purchase Amount (Normal)
purchase_amount = np.random.normal(loc=1500, scale=300, size=days)

# 3️⃣ Product Category Selection (Multinomial)
product_categories = np.random.multinomial(
    n=200,
    pvals=[0.3, 0.3, 0.2, 0.2],
    size=days
)

# 4️⃣ Delivery Time (Poisson)
delivery_time = np.random.poisson(lam=3, size=days)

# 5️⃣ Waiting Time Between Orders (Exponential)
waiting_time = np.random.exponential(scale=1/10, size=days)

# 6️⃣ Product Quality (Binomial)
product_quality = np.random.binomial(n=1, p=0.9, size=days)

print("Daily Customers:", daily_customers)
print("Purchase Amount:", purchase_amount)
print("Product Categories (A,B,C,D):", product_categories)
print("Delivery Time:", delivery_time)
print("Waiting Time:", waiting_time)
print("Product Quality:", product_quality)
```
