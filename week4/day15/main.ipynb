```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sns.set_style("whitegrid")
np.random.seed(42)
```

```python
dates = pd.date_range(start="2019-01-01", periods=60, freq="ME")

data = pd.DataFrame({
    "Date": dates,
    "Region": np.random.choice(["North","South","East","West"], 60),
    "Category": np.random.choice(["Electronics","Clothing","Home","Sports"], 60),
    "Sales": np.random.normal(50000, 10000, 60).round(2),
    "Profit": np.random.normal(8000, 2000, 60).round(2),
    "Discount": np.random.uniform(0,0.3,60).round(2)
})

data["Year"] = data["Date"].dt.year
data["Month"] = data["Date"].dt.month
```

```python
data.head()
data.info()
data.describe()
data.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Date", y="Sales", data=data)
plt.title("Sales Trend Over Time")
plt.show()
```

```python
sns.barplot(x="Region", y="Sales", data=data)
plt.title("Sales by Region")
plt.show()
```

```python
sns.histplot(data["Sales"], kde=True)
plt.title("Sales Distribution")
plt.show()
```

```python
corr = data.corr(numeric_only=True)

plt.figure(figsize=(8,5))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
data["Sales"].mean()
data["Sales"].median()
data["Sales"].std()
```
