```python
Feature engineering:
it is the process of finding the unique attribute of the data that can be considered as a feature
and also called as attribute
 
for example if we are performing a churn prediction the data inclues the features like
customer id
age
income
level
location
```

```python
it is the process to select,extract and to transform the variables  when creating a predicting modelling
```

```python
Feature engineering consists of different types like:
Feature creation
Feature transformation
Feature extraction
Feature selection
```

```python
Feature creation:

it refers to the creation of new features from the existing data to assist with better predictions
```

```python
Feature transformation:
it includes the steps for replacing the missing features that are not valid

some of the techniques includes 
one-hot encoding
non-linear transformations
```

```python
Feature extraction:
it involves reducing the amount of data to be process by using some of the techniques like
principle component analysis
independent component analysis
linear component analysis(reduce the dimensions)
```

```python
Feature selection:
it is the process of selecting a subset of extracted features
this uses to contribute to minimize the error rate of the trained model
```

```python
#train_test_Split
from sklearn.model_selection import train_test_split
x_train,y_train,x_test,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#regression--->linear regression

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
#ridge
from sklearn.linear_model import Ridge
model=Ridge(alpha=1.0)
model.fit(x_train,y_train)
```

```python
#lasso
from sklearn.linear_model import Lasso
model=Lasso(alpha=0.1)
model.fit(x_train,y_train)
```

```python
#polynomial
from sklearn.preprocessing import PolynomialFeatures
model=PolynomialFeatures(degree=3)
model.fit_transform(x)
```

```python
#classification---->logistic regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression
model.fit(x_train,y_train)
```

```python
#k-nearest neighbors
from sklearn.neighbors import KneighborsClassifier
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
```

```python
#support vector machine
from sklearn.svm import SVC
model=SVC(kernel='rbf')
model.fit(x_train,y_train)
```

```python
#decision tree classifier
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
#random forest classifier
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
#clustering--->Kmeans

from sklearn.cluster import KMeans
model=KMeans(n_clusters=3)
model.fit(x)
labels=model.predict(x)
```

```python
#dbscan
from sklearn.cluster import DBSCAN
model=DBSCAN(eps=0.5,min_samples=5)
labels=model.predict(x)
```

```python
#gmm(gaussian mixture models)
from sklearn.mixture import GaussianMixture
model=GaussianMixture(n_components=3)
model.fit(x)
labels=model.predict(x)
```

```python
#hiearchical
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering()
labels=model.fit_predict(x)
```
