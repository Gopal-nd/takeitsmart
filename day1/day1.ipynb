```python
import pandas as pd
data = {
    'name':["gopal","Reddy",None,"superman","goapl"],
    'age':[20,23,99,None,99],
    'work':['remort',None,'user man','stap','so']
}
df = pd.DataFrame(data)

df.to_csv("raw_logs.csv", index=False)
print("Raw log file created")
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
print(df)

# handle missing values
df['age'] = df['age'].fillna(df['age'].mean())
df['name'] = df['name'].fillna('unknown')
df['work'] = df['work'].fillna('unknown')

# clean text
df['name'] = df['name'].str.strip().str.lower()
df['work'] = df['work'].str.strip().str.lower()

# fix datatype
df['age'] = df['age'].astype(int)
```

```python
df.to_csv("cleaned_raw_logs.csv", index=False)
print("Cleaned raw logs saved successfully")
```
