```python
print("Internship Eligibility Checker using Dictionary")

student = {
    "cgpa": 7.5,
    "year": 3,
    "skill_score": 70
}

if student["cgpa"] >= 7 and student["year"] >= 2 and student["skill_score"] >= 60:
    print("Eligible")
else:
    print("Not eligible")
```

```python
print("E-Commerce Discount Engine")

order = {
    "cart_value": 6000,
    "membership": "gold"
}

if order["membership"] == "gold":
    discount = 0.2
elif order["cart_value"] >= 5000:
    discount = 0.1
else:
    discount = 0

print(order["cart_value"] - (order["cart_value"] * discount))
```

```python
print("Bank Account Balance Validation")

account = {
    "balance": 8000,
    "withdraw": 3000
}

if account["withdraw"] <= account["balance"]:
    print("Withdrawal allowed")
else:
    print("Insufficient balance")
```

```python
print("Attendance Status System")

attendance = ["present", "present", "absent", "present"]

if attendance.count("present") >= 3:
    print("Eligible")
else:
    print("Not eligible")
```

```python
print("Employee Attendance Counter")

employees = ["present", "absent", "present", "present"]
count = 0

for status in employees:
    if status == "present":
        count += 1

print(count)
```

```python
print("Internship Shortlisting System")

students = [
    {"name": "A", "cgpa": 8},
    {"name": "B", "cgpa": 6.5},
    {"name": "C", "cgpa": 7.2}
]

for s in students:
    if s["cgpa"] >= 7:
        print(s["name"])

```
