```python
print("1. Student Class with Objects")

class Student:
    def __init__(self, name, rollno, marks):
        self.name = name
        self.rollno = rollno
        self.marks = marks

s1 = Student("Gopal", 1, 85)
s2 = Student("Rahul", 2, 78)
s3 = Student("Anita", 3, 92)

for s in [s1, s2, s3]:
    print(s.name, s.rollno, s.marks)
```

```python
print("4. Rectangle Class")

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

r = Rectangle(10, 5)
print(r.area(), r.perimeter())
```

```python
print("6. Bank Account with Private Balance")

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance

acc = BankAccount(5000)
acc.deposit(2000)
acc.withdraw(1000)
print(acc.get_balance())
```

```python
print("7. User Class with Private Password")

class User:
    def __init__(self, password):
        self.__password = password

    def set_password(self, new_password):
        self.__password = new_password
    def get_password(self):
        return self.__password

u = User("1234")
u.set_password("abcd")
print("Password updated")
print(u.get_password())
```

```python
print("8. ATM Class with PIN Validation")

class ATM:
    def __init__(self, pin, balance):
        self.__pin = pin
        self.balance = balance

    def withdraw(self, entered_pin, amount):
        if entered_pin == self.__pin:
            if amount <= self.balance:
                self.balance -= amount
                return self.balance
        return "Transaction Failed ,Enter currect Password"

atm = ATM(1234, 5000)
try:
    print(atm.withdraw(1234, 2000))
    print(atm.withdraw(8234, 20))
except err:
    print(err)

```

```python
print("11. Account and Saving Account")

class Account:
    def __init__(self, balance):
        self.balance = balance

class SavingAccount(Account):
    def interest(self):
        return self.balance * 0.05

s = SavingAccount(10000)
print(s.interest())
```

```python
print("12. Person and Employee Inheritance")

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Employee(Person):
    def __init__(self, name, age, emp_id):
        super().__init__(name, age)
        self.emp_id = emp_id

e = Employee("Gopal", 22, 101)
print(e.name, e.age, e.emp_id)
```

```python
print("13. Shape Inheritance")

class Shape:
    pass

class Circle(Shape):
    def area(self, r):
        return 3.14 * r * r

class Rectangle(Shape):
    def area(self, l, w):
        return l * w

print(Circle().area(5))
print(Rectangle().area(4, 6))
```

```python
print("15. Method Overriding Salary")

class Employee:
    def calculate_salary(self):
        return 30000

class Manager(Employee):
    def calculate_salary(self):
        return 60000

class Intern(Employee):
    def calculate_salary(self):
        return 15000
print(Employee().calculate_salary())
print(Manager().calculate_salary())
print(Intern().calculate_salary())
```

```python
print("19. Payment System")

class Payment:
    def pay(self):
        pass

class EMI(Payment):
    def pay(self):
        print("Paid via EMI")

class CreditCard(Payment):
    def pay(self):
        print("Paid via Credit Card")

class DebitCard(Payment):
    def pay(self):
        print("Paid via Debit Card")

class Cash(Payment):
    def pay(self):
        print("Paid via Cash")

class UPI(Payment):
    def pay(self):
        print("Paid via UPI")

for p in [EMI(), CreditCard(), DebitCard(), Cash(), UPI()]:
    p.pay()
```

```python
print("Mini Project 1: Library Management System")

class Book:
    def __init__(self, name):
        self.name = name

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def view_books(self):
        for b in self.books:
            print(b.name)

lib = Library()
lib.add_book(Book("Python"))
lib.add_book(Book("DSA"))
lib.view_books()
```

```python
print("Mini Project 2: Student Result System")

class Student:
    def __init__(self, marks):
        self.marks = marks

    def grade(self):
        if self.marks >= 75:
            return "A"
        elif self.marks >= 60:
            return "B"
        else:
            return "Fail"

s = Student(82)
print(s.grade())
```

```python
print("Mini Project 3: E-Commerce Product System")

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class Cart:
    def __init__(self):
        self.items = []

    def add(self, product):
        self.items.append(product)

    def bill(self):
        print(sum(p.price for p in self.items))

cart = Cart()
cart.add(Product("Phone", 20000))
cart.add(Product("Headset", 2000))
cart.bill()
```

```python
print("Mini Project 4: Employee Management System")

class Employee:
    def __init__(self, name, salary):
        self._salary = salary
        self.name = name

    def calculate_salary(self):
        return self._salary

class Manager(Employee):
    def calculate_salary(self):
        return self._salary + 20000

class Intern(Employee):
    def calculate_salary(self):
        return self._salary

employees = [Manager("A", 50000), Intern("B", 15000)]

for e in employees:
    print(e.name, e.calculate_salary())
```

```python
print("Online Shopping System")

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class Electronics(Product):
    pass

class Clothing(Product):
    pass

class Cart:
    def __init__(self):
        self.items = []

    def add(self, item):
        self.items.append(item)

    def generate_bill(self):
        total = sum(i.price for i in self.items)
        print(total)

cart = Cart()
cart.add(Electronics("Laptop", 60000))
cart.add(Clothing("T-Shirt", 1500))
cart.generate_bill()
```
