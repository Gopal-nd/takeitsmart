```python
print("Task 01: Read CSV Data and Filter High Salary Employees")

import csv

high_salary = []

with open("employees.csv") as f:
    reader = csv.DictReader(f)
    for row in reader:
        if int(row["salary"]) >= 50000:
            high_salary.append(row)

print(high_salary)
```

```python
print("Task 02: Count Error Lines and Store Separately")

error_lines = []
count = 0

with open("log.txt") as f:
    for line in f:
        if "ERROR" in line:
            count += 1
            error_lines.append(line)

print(count)
print(error_lines)
```

```python
print("Custom ATM Withdrawal Exception")

class InsufficientBalance(Exception):
    pass

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientBalance("Insufficient balance")
    return balance - amount

try:
    print(withdraw(3000, 5000))
except InsufficientBalance as e:
    print(e)
```

```python
print("Project 2: Word Counter Tool")

try:
    with open("summary.txt") as f:
        lines = f.readlines()
        words = sum(len(line.split()) for line in lines)
        chars = sum(len(line) for line in lines)
        print(len(lines), words, chars)
except FileNotFoundError:
    print("File not found")
```

```python
print("Project 3: Employee Salary Analyzer")

import csv

salaries = []

try:
    with open("employee.csv") as f:
        reader = csv.DictReader(f)
        for row in reader:
            salaries.append(int(row["salary"]))
    print(max(salaries))
    print(min(salaries))
    print(sum(salaries) / len(salaries))
except FileNotFoundError:
    print("File not found")
```

```python
print("Project 4: ATM Simulation")

class ATM:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            return self.balance
        else:
            raise Exception("Insufficient balance")

atm = ATM(5000)

try:
    print(atm.withdraw(2000))
except Exception as e:
    print(e)
```

```python
print("Project 5: User Login System")

def login(username):
    try:
        with open("user.txt") as f:
            users = f.read().splitlines()
            if username in users:
                print("Login successful")
            else:
                print("Invalid user")
    except FileNotFoundError:
        print("User file not found")

login("admin")
```

```python
print("Automated Report Generator Use Case")

def generate_report(filename):
    try:
        with open(filename) as f:
            data = f.read()
        report = f"Report Generated Successfully\n{data}"
        with open("auto_report.txt", "w") as r:
            r.write(report)
        print("Report created")
    except FileNotFoundError:
        print("Source file missing")

generate_report("summary.txt")
```
