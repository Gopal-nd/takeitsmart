```python
print("Task 01: Student Grade Calculator")

def grade_calculator(marks):
    avg = sum(marks) / len(marks)
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 60:
        return "C"
    else:
        return "Fail"

print(grade_calculator([78, 85, 90, 72]))
```

```python
print("Task 02: ATM Withdrawal Function")

def withdraw(balance, amount):
    if amount <= balance and amount % 100 == 0:
        return balance - amount
    return "Invalid transaction"

print(withdraw(5000, 1200))

```

```python
print("Task 03: Password Validator")

def password_validator(password):
    if len(password) >= 8 and any(c.isdigit() for c in password) and any(c.isupper() for c in password):
        return "Valid Password"
    return "Invalid Password"

print(password_validator("Secure123"))
```

```python
print("Task 04: Employee Salary Calculator")

def salary_calculator(basic):
    hra = basic * 0.20
    pf = basic * 0.12
    return basic + hra - pf

print(salary_calculator(30000))
```

```python
print("Task 05: File Data Counter")

def file_counter(filename):
    with open(filename) as f:
        lines = f.readlines()
    words = sum(len(line.split()) for line in lines)
    characters = sum(len(line) for line in lines)
    return len(lines), words, characters

print(file_counter('data.txt'))
```

```python
print("Even or Odd Checker")

def even_or_odd(n):
    return "Even" if n % 2 == 0 else "Odd"

print(even_or_odd(7))

```

```python
print("Sum from 1 to n")

def sum_n(n):
    return sum(range(1, n + 1))

print(sum_n(10))
```

```python
print("Celsius to Fahrenheit using Lambda")

c_to_f = lambda c: (c * 9/5) + 32
print(c_to_f(30))
```

```python
print("Remove Negative Numbers using Lambda and Filter")

numbers = [-5, 3, -1, 7, 0]
result = list(filter(lambda x: x >= 0, numbers))
print(result)
```

```python
print("Factorial of 5")

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    print(result)

factorial(5)
```

```python
print("Reverse Each Word in a Sentence")

def reverse_words(sentence):
    return " ".join(word[::-1] for word in sentence.split())

print(reverse_words("hello world python"))
```

```python
print("Frequency of Elements")

def frequency_counter(lst):
    freq = {}
    for i in lst:
        freq[i] = freq.get(i, 0) + 1
    return freq

print(frequency_counter([1, 2, 2, 3, 1, 1]))

```

```python
print("Prime Number Checker")

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(11))
```

```python
print("Pass or Fail using Lambda")

percentage = 90
result = lambda p: "Pass" if p >= 40 else "Fail"
print(result(percentage))

```

```python
print("Print Vowels from String")

def vowels_from_string(s):
    return [c for c in s if c.lower() in "aeiou"]

print(vowels_from_string("Internship"))
```

```python
print("Task 01: Even Number Filter using Lambda")

nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
# filter(lambda x:x%2==0,[1,2,3,4,5])
print(evens)
```

```python
print("Task 02: Sort Employees by Salary using Lambda")

employees = [
    {"name": "A", "salary": 30000},
    {"name": "B", "salary": 50000},
    {"name": "C", "salary": 40000}
]

employees.sort(key=lambda x: x["salary"])
print(employees)
```

```python
print("Task 03: Square and Cube using Lambda and Map")

nums = [2, 3, 4]
squares = list(map(lambda x: x**2, nums))
cubes = list(map(lambda x: x**3, nums))
print(squares, cubes)
```

```python
print("Task 04: Discount Calculator using Lambda")

discount = lambda price: price * 0.9 if price >= 5000 else price
print(discount(6000))
```
