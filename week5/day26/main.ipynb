```python

import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

np.random.seed(10)

days = np.arange(1, 101)
actual = np.cumsum(np.random.normal(2, 5, 100)) + 200
predicted = actual + np.random.normal(0, 3, 100)
volume = np.random.randint(1000, 5000, 100)

monthly_returns = np.array_split(actual, 5)
monthly_avg = [m.mean() for m in monthly_returns]

fig = make_subplots(
    rows=3, cols=2,
    specs=[[{"colspan":2}, None],
           [{"type":"scatter"},{"type":"bar"}],
           [{"type":"pie"},{"type":"indicator"}]],
    subplot_titles=("Actual vs Predicted Price",
                    "Volume vs Price",
                    "Monthly Avg Price",
                    "Sector Distribution",
                    "Model Accuracy")
)

fig.add_trace(go.Scatter(x=days,y=actual,name="Actual"),row=1,col=1)
fig.add_trace(go.Scatter(x=days,y=predicted,name="Predicted"),row=1,col=1)

fig.add_trace(go.Scatter(x=volume,y=actual,mode="markers",name="Volume"),row=2,col=1)

fig.add_trace(go.Bar(x=["M1","M2","M3","M4","M5"],y=monthly_avg,name="Monthly"),row=2,col=2)

fig.add_trace(go.Pie(labels=["IT","Banking","Pharma"],values=[40,35,25],hole=0.4),row=3,col=1)

fig.add_trace(go.Indicator(
    mode="gauge+number",
    value=92,
    title={'text':"Accuracy %"},
    gauge={'axis':{'range':[0,100]}}
),row=3,col=2)

fig.update_xaxes(rangeslider_visible=True,row=1,col=1)

fig.update_layout(height=850,width=850,title="Stock Prediction Dashboard",template="plotly_dark")
fig.show()


np.random.seed(20)

days = np.arange(1,101)
revenue = np.cumsum(np.random.randint(100,300,100))
profit = revenue - np.random.randint(500,2000,100)

regions = ["North","South","East","West"]
region_sales = np.random.randint(2000,5000,4)

fig = make_subplots(
    rows=3, cols=2,
    specs=[[{"colspan":2}, None],
           [{"type":"scatter"},{"type":"bar"}],
           [{"type":"pie"},{"type":"indicator"}]],
    subplot_titles=("Revenue vs Profit",
                    "Customer Growth",
                    "Region Sales",
                    "Product Share",
                    "Target Achievement")
)

fig.add_trace(go.Scatter(x=days,y=revenue,name="Revenue"),row=1,col=1)
fig.add_trace(go.Scatter(x=days,y=profit,name="Profit"),row=1,col=1)

fig.add_trace(go.Scatter(x=days,y=np.random.randint(20,80,100),mode="markers",name="Customers"),row=2,col=1)

fig.add_trace(go.Bar(x=regions,y=region_sales),row=2,col=2)

fig.add_trace(go.Pie(labels=["A","B","C"],values=[50,30,20],hole=0.4),row=3,col=1)

fig.add_trace(go.Indicator(
    mode="gauge+number",
    value=78,
    title={'text':"Target %"},
    gauge={'axis':{'range':[0,100]}}
),row=3,col=2)

fig.update_xaxes(rangeslider_visible=True,row=1,col=1)
fig.update_layout(height=850,width=850,title="Sales Dashboard",template="plotly_dark")
fig.show()
     


np.random.seed(30)

days = np.arange(1,101)
rides = np.random.randint(200,800,100)
duration = np.random.randint(5,45,100)

hourly = np.random.randint(50,200,24)

fig = make_subplots(
    rows=3, cols=2,
    specs=[[{"colspan":2}, None],
           [{"type":"scatter"},{"type":"bar"}],
           [{"type":"pie"},{"type":"indicator"}]],
    subplot_titles=("Ride Demand Trend",
                    "Peak Duration",
                    "Hourly Trips",
                    "Ride Type",
                    "Avg Duration")
)

fig.add_trace(go.Scatter(x=days,y=rides,name="Trips"),row=1,col=1)

fig.add_trace(go.Scatter(x=days,y=duration,mode="markers",name="Duration"),row=2,col=1)

fig.add_trace(go.Bar(x=list(range(24)),y=hourly),row=2,col=2)

fig.add_trace(go.Pie(labels=["UberX","Premier","Auto"],values=[60,25,15],hole=0.4),row=3,col=1)

fig.add_trace(go.Indicator(
    mode="gauge+number",
    value=np.mean(duration),
    title={'text':"Avg Trip Time"},
    gauge={'axis':{'range':[0,60]}}
),row=3,col=2)

fig.update_xaxes(rangeslider_visible=True,row=1,col=1)
fig.update_layout(height=850,width=850,title="Uber Dashboard",template="plotly_dark")
fig.show()
     


     

np.random.seed(40)

days = np.arange(1,101)
orders = np.random.randint(100,500,100)
delivery_time = np.random.randint(10,60,100)

categories = ["Grocery","Snacks","Beverages","Personal Care"]
sales = np.random.randint(2000,7000,4)

fig = make_subplots(
    rows=3, cols=2,
    specs=[[{"colspan":2}, None],
           [{"type":"scatter"},{"type":"bar"}],
           [{"type":"pie"},{"type":"indicator"}]],
    subplot_titles=("Order Trend",
                    "Delivery Time",
                    "Category Sales",
                    "Product Share",
                    "Avg Delivery Time")
)

fig.add_trace(go.Scatter(x=days,y=orders,name="Orders"),row=1,col=1)

fig.add_trace(go.Scatter(x=days,y=delivery_time,mode="markers"),row=2,col=1)

fig.add_trace(go.Bar(x=categories,y=sales),row=2,col=2)

fig.add_trace(go.Pie(labels=["Daily","Weekly","Monthly"],values=[50,30,20],hole=0.4),row=3,col=1)

fig.add_trace(go.Indicator(
    mode="gauge+number",
    value=np.mean(delivery_time),
    title={'text':"Avg Delivery"},
    gauge={'axis':{'range':[0,60]}}
),row=3,col=2)

fig.update_xaxes(rangeslider_visible=True,row=1,col=1)
fig.update_layout(height=850,width=850,title="Blinkit Dashboard",template="plotly_dark")
fig.show()
     


     
```
